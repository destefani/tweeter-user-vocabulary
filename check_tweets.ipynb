## Data Processing

### Datetime

```python
df['created_at'] = pd.to_datetime(df['created_at'])

df['year'] = df['created_at'].apply(lambda x: x.year)
df['month'] = df['created_at'].apply(lambda x: x.month)
df['week'] = df['created_at'].apply(lambda x: x.week)
```

## Text processing pipeline

- Remove stopwords

### 1. Lower Case

```python
example_text = example_text.lower()
example_text
```

2. List 

```python
example_text = example_text.split()
example_text
```

3. Count

```python
import itertools

all_words =list(itertools.chain(*splited_tweets))
word_count = collections.Counter(all_words)
```

```python
import nltk
from nltk.corpus import stopwords

stop_words = set(stopwords.words('spanish'))
```

```python
all_tweets = list(df['text'])
lower_tweets = [tweet.lower() for tweet in all_tweets]
splited_tweets = [tweet.split() for tweet in lower_tweets]
all_words = list(itertools.chain(*splited_tweets))
all_words = [word for word in all_words if not word in stop_words]
word_count = collections.Counter(all_words)

word_count.most_common()
```
